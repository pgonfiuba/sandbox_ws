# Angulos de Euler rpy

- De ángulos rpy a matriz de rotación (problema directo)
- De matriz de rotación a ángulos rpy (problema inverso)

```python
import numpy as np
from scipy.spatial.transform import Rotation as R

# Matriz de rotación (ejemplo)
R_mat = R.from_euler('xyz', [0.3, 0.5, 1.0]).as_matrix()
print("Matriz de rotación: \n",R_mat)

# A Euler (roll, pitch, yaw)
rpy = R.from_matrix(R_mat).as_euler('xyz')
print("RPY:", rpy)

# Volver a matriz
R_mat2 = R.from_euler('xyz', rpy).as_matrix()
print("Vuelvo a matriz de rotación: \n",R_mat2)
```

## Dos soluciones posibles

```python
rpy1 = np.array([0.3, 2.0, 1.0])
R_mat = R.from_euler('xyz', rpy1).as_matrix()
print("Primera solución rpy =", rpy1)
print("Matriz de rotación R=\n",R_mat)

# otra representación válida
rpy2 = R.from_matrix(R_mat).as_euler('xyz')
R_mat2 = R.from_euler('xyz', rpy1).as_matrix()
print("Segunda solución rpy =", rpy2)
print("Matriz de rotación R=\n",R_mat2)
```

## Gimbal lock

```python
# pitch = 90° → gimbal lock
# Ejemplo de slides
rpy_gl = np.array([0.0, -np.pi/2, np.pi/2])
R_mat = R.from_euler('xyz', rpy_gl)

# Intentar recuperar Euler
rpy_rec = R_mat.as_euler('xyz')

print("RPY original:", rpy_gl)
print("RPY recuperado:", rpy_rec)

# Tengo infinitas soluciones: en este ejemplo cualquier combinación r+y=pi/2
```
